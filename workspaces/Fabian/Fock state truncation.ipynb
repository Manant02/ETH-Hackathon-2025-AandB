```python
import dynamiqs as dq
import jax.numpy as jnp
import jax
from dynamiqs import *
import matplotlib.pyplot as plt
from dynamiqs import *

```

```python
N=30
alpha=2
```

```python
cat_state=1/jnp.sqrt(2)*(coherent(N,alpha)+coherent(N,-alpha))
xvec, yvec, W_cat=wigner(cat_state, 5,5)
plt.contourf(xvec, yvec,W_cat.T, levels=100,cmap="seismic", vmax=2/jnp.pi, vmin=-2/jnp.pi)
plt.xlabel("x")
plt.ylabel("p")
plt.colorbar()

```

```python
for N in [4,6,10,15,30]:
    alpha=2
    cat_state=1/jnp.sqrt(2)*(coherent(N,alpha)+coherent(N,-alpha))
    xvec, yvec, W_cat=wigner(cat_state, 5,5)
    
    

    plt.plot(xvec,W_cat[100], label=f"N={N}")
    plt.legend()

plt.xlabel('p')
plt.ylabel('W(0,p)')
```
